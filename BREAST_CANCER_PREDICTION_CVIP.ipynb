## Find the Outliers by box plot

```python
plt.figure(figsize=(20,20))
plotnumber=1
for col in cancer_data.columns:
  if plotnumber <= 25:
    ax=plt.subplot(5,5,plotnumber)
    sns.boxplot(cancer_data[col])
    plt.xlabel(col, fontsize=15)
  plotnumber+=1
plt.tight_layout()
plt.show()
```

## Following the Interquartile Formule to the remove Outliers

```python
for col in cancer_data.columns:
  a=cancer_data[col].quantile(0.25)
  b=cancer_data[col].quantile(0.75)
  iqr=b-a
  upper_limit=b+(1.5*iqr)
  lower_limit=a-(1.5*iqr)
  cancer_data[col] = np.where(cancer_data[col] < lower_limit, lower_limit,
  np.where(cancer_data[col] > upper_limit, upper_limit,cancer_data[col]))
```

## Boxplot of the dataset after removing the outliers

```python
plt.figure(figsize=(20,15))
plotnumber=1
for col in cancer_data.columns:
  if plotnumber <= 25:
    ax=plt.subplot(5,5,plotnumber)
    sns.boxplot(cancer_data[col])
    plt.xlabel(col, fontsize=15)
  plotnumber+=1
plt.tight_layout()
plt.show()
```

## Spliting the x and y values

```python
x=cancer_data.drop(['diagnosis'],axis=1)
y=cancer_data.diagnosis
```

## Import train test and split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

## Import standard scaler

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler()
x_train=standard_scaler.fit_transform(x_train)
x_test=standard_scaler.transform(x_test)
```

## Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

## import the algorithms needed

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

## Algorithms

```python
algorithms = ['DecisionTreeClassifier','GaussianNB','LogisticRegression','KNeighborsClassifier','RandomForestClassifier','SVC','AdaBoostClassifier']
value = [ DecisionTreeClassifier(criterion="entropy",random_state=42),GaussianNB(),LogisticRegression(), KNeighborsClassifier(n_neighbors = 9, weights ='uniform'),RandomForestClassifier(n_estimators=300), SVC(kernel="rbf",random_state=15), AdaBoostClassifier(n_estimators=100)]
models = dict(zip(algorithms,value))
models
```

## Predicted value and Accuracy score

```python
predicted =[]
for name,algo in models.items():
    model=algo
    model.fit(x_train,y_train)
    predict = model.predict(x_test)
    accuracy = accuracy_score(y_test, predict)*100
    predicted.append(accuracy)
    print(name,accuracy)
```

## Plot the Accuracy score for the difne algorithms

```python
plt.figure(figsize = (10,10))
sns.barplot(x = predicted, y = algorithms, palette='magma')
plt.title("ACCURACY VALUES", fontsize=20)
```

# Analysis only necessary columns 

```python
data=data[['eventid','iyear','imonth','iday','extended','country','country_txt','region','targsubtype1','attacktype1',
           'latitude','longitude','specificity','INT_IDEO','INT_MISC','INT_ANY']]
```

# remaining column name

```python
data=data.rename(columns={'iyear':'Year','imonth':'Month', 'iday':'Day','extended':'Extended','country':'Country','country_txt':'Country in text','region':'Region','targsubtype1':'Targsub',
                          'attacktype1':'Attacks','latitude':'Lat','longitude':'Lon','specificity':'Specific','INT_IDEO':'Intideo',
       'INT_MISC':'Intmisc', 'INT_ANY':'Intany'})
```

```python
data.columns
```

```python
data
```

# FILLING UP THE NULL VALUES

```python
data['Intideo']=data['Intideo'].fillna('Unknown')
data['Intmisc']=data['Intmisc'].fillna('Unknown')
data['Intany']=data['Intany'].fillna('Unknown')
data['Lat']=data['Lat'].fillna('Unknown')
data['Lon']=data['Lon'].fillna('Unknown')
data['Targsub']=data['Targsub'].fillna('Unknown')
     
```

```python
data.isnull().sum()
```

# Visualization use plotting

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Day','Attacks In Month']
columns=['Day','Month']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i])
  sns.kdeplot(x=data[col])
plt.tight_layout()
plt.show()

```

# Attacks in the day and month

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Day','Attacks In Month']
columns=['Day','Month']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i])
  sns.stripplot(x=data[col])
plt.tight_layout()
plt.show()
```

# Attacks in the region and country

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Region','Attacks In Country']
columns=['Region','Country']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i],fontsize=20)
  plt.xticks(rotation=360)
  plt.xlabel(col,fontsize=20)
  sns.countplot(x=data[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Region','Attacks In Country']
columns=['Region','Country in text']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i],fontsize=20)
  plt.xticks(rotation=90)
  plt.xlabel(col,fontsize=20)
  sns.countplot(x=data[col])
plt.tight_layout()
plt.show()
```

# Vinolin plot for attacks

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.title('Attacks')
sns.violinplot(x=data['Attacks'],y=data['Day'])
```

# Attacks in the each region based on the year

```python
pd.crosstab(data.Year,data.Region).plot(kind='area',figsize=(15,8))
plt.title("ATTACKS IN EACH REGION")
plt.xlabel('Year')
plt.ylabel('Attacks')
plt.show()
```
